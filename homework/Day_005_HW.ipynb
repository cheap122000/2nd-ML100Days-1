```python
print("MEAN:AMT_INCOME_TOTAL→",app_train['AMT_INCOME_TOTAL'].mean())
print("MEAN:AMT_CREDIT→",app_train['AMT_CREDIT'].mean())
print("MEAN:AMT_ANNUITY→",app_train['AMT_ANNUITY'].mean())
print("MEAN:AMT_GOODS_PRICE→",app_train['AMT_GOODS_PRICE'].mean())
```

```python
print("STD:AMT_INCOME_TOTAL→",app_train['AMT_INCOME_TOTAL'].std())
print("STD:AMT_CREDIT→",app_train['AMT_CREDIT'].std())
print("STD:AMT_ANNUITY→",app_train['AMT_ANNUITY'].std())
print("STD:AMT_GOODS_PRICE→",app_train['AMT_GOODS_PRICE'].std())
```

```python
#.describe()可以看相關的敘述統計
print("DESCRIBE:AMT_INCOME_TOTAL→",app_train['AMT_INCOME_TOTAL'].describe())
print("---"*10)
print("DESCRIBE:AMT_CREDIT→",app_train['AMT_CREDIT'].describe())
print("---"*10)
print("DESCRIBE:AMT_ANNUITY→",app_train['AMT_ANNUITY'].describe())
print("---"*10)
print("DESCRIBE:AMT_GOODS_PRICE→",app_train['AMT_GOODS_PRICE'].describe())
```

```python
#app_train['AMT_INCOME_TOTAL']
#app_train['AMT_CREDIT']
#app_train['AMT_ANNUITY']
#app_train['AMT_GOODS_PRICE']

#bins: 直方圖的柱數，默認為10
#edgecolor: 直方圖邊框顏色
#alpha: 透明度
#histtype: 直方圖類型
#subplot(nrows, ncols, index, **kwargs)
#figsize：圖表長寬

#plt.figure(figsize = (15,20))
plt.subplots(4,1,figsize=(15,20))
AMT_INCOME_TOTAL = plt.subplot(4, 1, 1)
AMT_INCOME_TOTAL = app_train['AMT_INCOME_TOTAL'].hist(bins=30,color="LIMEGREEN",edgecolor="BLACK",alpha=0.7,histtype='bar',range = (20000,150000))
AMT_INCOME_TOTAL.set_xlabel("AMT_INCOME_TOTAL")
AMT_INCOME_TOTAL.set_ylabel("Frequency")

AMT_CREDIT = plt.subplot(4, 1, 2)
AMT_CREDIT = app_train['AMT_CREDIT'].hist(bins=30,color="HOTPINK",edgecolor="BLACK",alpha=0.7,histtype='bar',range = (40000,150000))
AMT_CREDIT.set_xlabel("AMT_CREDIT")
AMT_CREDIT.set_ylabel("Frequency")

AMT_ANNUITY = plt.subplot(4, 1, 3)
AMT_ANNUITY = app_train['AMT_ANNUITY'].hist(bins=30,color="DEEPSKYBLUE",edgecolor="BLACK",alpha=0.7,histtype='bar',range = (0,100000))
AMT_ANNUITY.set_xlabel("AMT_ANNUITY")
AMT_ANNUITY.set_ylabel("Frequency")

AMT_GOODS_PRICE = plt.subplot(4, 1, 4)
AMT_GOODS_PRICE = app_train['AMT_GOODS_PRICE'].hist(bins=30,color="BROWN",edgecolor="BLACK",alpha=0.5,histtype='bar',range = (40000,150000))
AMT_GOODS_PRICE.set_xlabel("AMT_GOODS_PRICE")
AMT_GOODS_PRICE.set_ylabel("Frequency")
```
